# **IQ Data Intuition and Mechanics**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_data(range, frequencies, points):
    Qs = np.random.randint(low=range[0], high=range[1], size=points) # Sin Function
    Is = np.random.randint(low=range[0], high=range[1], size=points) # Cos Function
    print(f"Sin Amplitude Modulations: {Qs}")
    print(f"Cos Amplitude Modulations: {Is}")
    time_steps = np.linspace(start=0, stop=points, num=points)

    sin_freq = frequencies[0]
    cos_freq = frequencies[1]

    sin_data = np.sin(2 * np.pi * time_steps * sin_freq)
    cos_data = np.cos(2 * np.pi * time_steps * cos_freq)

    sin_amp_modulation = np.multiply(sin_data, Qs)
    cos_amp_modulation = np.multiply(cos_data, Is)

    return sin_amp_modulation, cos_amp_modulation
```

```python
range = (1, 8)
frequencies = np.random.randint(low=1, high=10, size=2)
points = 100000

steps = np.linspace(start=0, stop=points, num=points)
I, Q = generate_data(range=range, frequencies=frequencies, points=points)
Combined_IQ = I + Q

print(f"Range of Modulations: {range}")
print(f"Frequencies of Sin and Cos Waves: {frequencies}")
print(f"Number of Points Sampled: {points}")

print(f"I Data (0: 10): {I[0:11]}")
print(f"Q Data (0: 10): {Q[0:11]}")
```

```python
plt.figure(figsize=(50, 10))
plt.title("Sin & Cos & Combined: Amplitude vs Time")
plt.scatter(steps, I, alpha=0.8)
plt.scatter(steps, Q, alpha=0.8)
plt.plot(steps, Combined_IQ, alpha=0.6)
plt.legend()
plt.grid()
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

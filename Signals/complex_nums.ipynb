# **Complex Number Intuitions and Implementations**

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
import seaborn as sns

import pygame
```

```python
def rotate(num, rotation_angle):
    result = num * complex(np.cos(rotation_angle), np.sin(rotation_angle))
    return result

# ~~~~~~~~~~~~~~~~~~
complex_num = 2 + 3j

rotation_angle = np.pi / 2
rotated = rotate(complex_num, rotation_angle)

plt.scatter(complex_num.real, complex_num.imag, color='blue', label="Original")
plt.scatter(rotated.real, rotated.imag, color='red', label=f"Rotated by {np.degrees(rotation_angle):.0f}Â°")

plt.xlabel("Real Part")
plt.ylabel("Imaginary Part")
plt.title("Plotting a Complex Number in the Complex Plane")
plt.grid(True)
plt.legend()
plt.show()
```

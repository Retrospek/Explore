# **Markov Chain Monte Carlo (MCMC)**

```python
import numpy as np
import math

import matplotlib.pyplot as plt
```

### **Intuition**

**Goal:** To utilize some algorithm to find high-yeild samples from a compelex distribution p(x), so that we can approximate it through samples
- Remember we are NOT trying to find the ACTUAL distribution, rather we want representative samples from such distribution

```python
def proposal(x):
    perturbation = np.random.normal(loc=0.0, scale=50)

    prop = x + perturbation

    return prop
```

```python
def accept_reject(x_hat, x):
    alpha = min(1, np.exp(-0.5 * (x_hat**2 - x**2))) # ratio of

    u = np.random.uniform(low=0, high=1)

    if u <= alpha:
        return True
    else:
        return False
```

```python
samples = [200]
for _ in range(20000):
    x = samples[-1]
    x_hat = proposal(x)

    accept = accept_reject(x_hat, x)

    if accept:
        samples.append(x_hat)
    else:
        samples.append(x)

samples = samples[5000:]
print(samples)
```
